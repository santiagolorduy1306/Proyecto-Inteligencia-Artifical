```python
#Unión de los archivos para obtener un solo dataset con todas las columnas y filas necesarias
```

```python
train_ = pd.merge(train,oil, how = "left", on = "date")
train_ = pd.merge(train_, HE, how = "left", on = "date")
train_ = pd.merge(train_, transactions, how = "left", on =["date","store_nbr"])
train_ = pd.merge(train_,stores, how = "left", on = "store_nbr")
train_.rename(columns={"type_x":"holiday_type", "type_y":"store_type"}, inplace = True)
train_.tail()
```

```python
test_ = pd.merge(test,oil, how = "left", on = "date")
test_ = pd.merge(test_, HE, how = "left", on = "date")
test_ = pd.merge(test_, transactions, how = "left", on =["date","store_nbr"])
test_ = pd.merge(test_,stores, how = "left", on = "store_nbr")
test_.rename(columns={"type_x":"holiday_type", "type_y":"store_type"}, inplace = True)
test_.tail()
```

```python
# cambiando los NaN por valores que puedan ser utilizados
```

```python
train_["dcoilwtico"] = train_["dcoilwtico"].fillna(method = "bfill")
train_["holiday_type"] = train_["holiday_type"].fillna(0)
train_["locale"] = train_["locale"].fillna(0)
train_["locale_name"] = train_["locale_name"].fillna(0)
train_["description"] = train_["locale"].fillna(0)
train_["transferred"] = train_["transferred"].fillna(0)
train_["transactions"] = train_["transactions"].fillna(0)
train_.isnull().sum()

```

```python
test_["dcoilwtico"] = test_["dcoilwtico"].fillna(method = "bfill")
test_["holiday_type"] = test_["holiday_type"].fillna(0)
test_["locale"] = test_["locale"].fillna(0)
test_["locale_name"] = test_["locale_name"].fillna(0)
test_["description"] = test_["locale"].fillna(0)
test_["transferred"] = test_["transferred"].fillna(0)
test_["transactions"] = test_["transactions"].fillna(0)
test_.isnull().sum()
```
